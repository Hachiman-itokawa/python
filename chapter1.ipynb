4. Logical Operators

```python
x = 5
print(x > 2 and x < 10)
```

5. Bitwise Operators

Operate on bits (0 and 1).

&(and)

|(or)

^(xor)

~(not)

<<(left shift)

>>(right shift)

```python

a, b = 6, 3   # 6 -> 110, 3 -> 011 (binary)

print("a & b =", a & b)   
print("a | b =", a | b)   
print("a ^ b =", a ^ b)  
print("~a =", ~a)         
print("a >> 1 =", a >> 1) 
print("a << 1 =", a << 1)
```

6.Membership Operators

in(present)

not in(not present)

```python
x = [1,2,3]
print(2 in x,4 not in x)
```

7.Identity Operators

Check if two variables refer to same object

is(same object)

is not(different object)

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a == b:", a == b)     
print("a is b:", a is b)     
print("a is c:", a is c)
```
