#### Q4. Check whether a number is Armstrong number.

```python
n=int(input("Enter a number: "))
temp=n
sum=0
while temp>0:
    digit=temp%10
    sum+=digit**3
    temp//=10
if sum==n:
    print(f"{n} is an Armstrong number.")
else:
    print(f"{n} is not an Armstrong number.")
```

#### Q6. Check whether a number is prime.

```python
n=int(input("Enter a number: "))
if n>1:
    for i in range(2,n):
        if n%i==0:
            print(f"{n} is not a prime number.")
            break
    else:
        print(f"{n} is a prime number.")
```

#### Q7. Check palindrome number.

```python
n=int(input("Enter a number: "))
temp=n
rev=0
while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp//=10
if rev==n:
    print(f"{n} is a palindrome number.")
else:
    print(f"{n} is not a palindrome number.")
```

#### Q8. Count uppercase, lowercase and digits.

```python
s=input("Enter a string: ")
upper=0
lower=0
digits=0
for c in s:
    if c.isupper():
        upper+=1
    elif c.islower():
        lower+=1
    elif c.isdigit():
        digits+=1
print(f"The number of uppercase letters in the string is: {upper}")
print(f"The number of lowercase letters in the string is: {lower}")
print(f"The number of digits in the string is: {digits}")
```
