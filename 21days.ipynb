#### Q3. Find factorial of a number.

```python
n=int(input("Enter a number: "))
fact=1
for i in range(1,n+1):
    fact*=i
print(f"Factorial of {n} is {fact}")
```

#### Q4. Check whether a number is Armstrong number.

```python
# Q4. Check whether a number is Armstrong number.
n=int(input("Enter a number: "))
temp=n
sum=0
while temp>0:
    digit=temp%10
    sum+=digit**3
    temp//=10
if sum==n:
    print(f"{n} is an Armstrong number.")
else:
    print(f"{n} is not an Armstrong number.")
```

#### Q6. Check whether a number is prime.

```python
# Check whether a number is prime.
n=int(input("Enter a number: "))
if n>1:
    for i in range(2,n):
        if n%i==0:
            print(f"{n} is not a prime number.")
            break
    else:
        print(f"{n} is a prime number.")
```

#### Q7. Check palindrome number.

```python
n=int(input("Enter a number: "))
temp=n
rev=0
while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp//=10
if rev==n:
    print(f"{n} is a palindrome number.")
else:
    print(f"{n} is not a palindrome number.")
```

#### Q10. Print all prime numbers between 1 to 50.

```python
for num in range(1,51):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num,end=" ")
```
