#### 1. Scan n values in range 0-3 and print the number of times each value has occurred. 

```python
n=int(input("enter a number: "))
freq=[0,0,0,0]

for i in range(n):
    num=int(input("enter a number(0-3): "))
    if 0<=num<=3:
        freq[num]+=1
for i in range(4):
    print(i,"occurred", freq[i],"times")
```

<h4> 3. WAP to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output.  <br><br>
Sample Input  <br>
N = 5  <br>
Scores= 2 3 6 6 5 <br><br>
Sample output  <br>
5  <br><br>
Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.</h4>

```python
n=int(input("enter a number: "))
li=[]

for i in range(n):
    total=int(input("fill the list:"))
    li.append(total)

print([li])
print("maximum no is ", max(li))
li.remove(max(li))
print([li])
```

<h4> 4. Create a dictionary of n persons where key is name and value is city.<br>   
a. Display all names <br> 
b. Display all city names  <br>
c. Display student names and cities of all students.<br>  
d. Count the number of students in each city. </h4>

```python
stu={}
n=int(input("enter a how many students number:"))
for i in range(1,n+1):
    student=input("enter student name:")
    city=input("enter city name:")
    stu[student]=city
print(stu)
print("\nStudents name are")
for name in stu.keys():
    print(name)
print("\nCities are")
for city in stu.values():
    print (city)
print("\nStudent and their cities are")
for name,city in stu.items():
    print(name,"lives in", city)

city_count={}
for city in stu.values():
    if city in city_count:
        city_count[city]+=1
    else:
        city_count[city]=1

print("\nNumber of students in each city")
for city,count in city_count.items():
    print(city,"has" ,count,"students")

```

<h4> 5. Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name,  year, director name, production cost, collection made (earning) & perform the following :- <br> 
a. print all movie details  <br>
b. display name of movies released before 2015  <br>
c. print movies that made a profit.  <br>
d. print movies directed by a particular director.</h4>

```python
movie={}
n=int(input("enter a number that how many movies you want to enter:"))
for i in range(1,n+1):
    name=input("enter movie name:")
    year=int(input("enter year in which year it released:"))
    director_name=input("enter director name:")
    producation_cost=int(input("enter production cost:"))
    collection=int(input("enter collection:"))
    movie[name]=(year, director_name, producation_cost, collection)
print("\nMovies and their details are:")
for name,details in movie.items():
    print(f"Movie: {name}")
    print(f"Year: {details[0]}")
    print(f"Director: {details[1]}")
    print(f"Production Cost: {details[2]}")
    print(f"Collection: {details[3]}")
    print()
print("\nMovies released before 2015:")
for name, details in movie.items():
    if details[0] < 2015:
        print(f"Movie: {name}")
        print(f"Year: {details[0]}")
        print()
print("\nMovies that made a profit:")
for name,details in movie.items():
    if details[3]>details[2]:
        print(f"Movie: {name}")
        print(f"Collection: {details[3]}")
        print(f"Production Cost: {details[2]}")
        print(f"Profit: {details[3]-details[2]}")
        print()
print ("\nMovies directed by a particular director:")
director = input("Enter the director's name: ")
for name, details in movie.items():
    if details[1] == director:
        print(f"Movie: {name}")
        print()
```
