#### 1.  Write a program to count and display the number of capital letters in a given string.

```python
a=input("enter a string:")
count=0
for ch in a:
    if ch.isupper():
        count+=1
print("total capital in it is ",count)
```

#### 2. Count total number of vowels in a given string. 

```python
m=input("enter a string:")
count=0
for ch in m.lower():
    if(ch=="a" or ch=="e" or ch=="i" or ch=="o" or ch=="u"):
        count+=1
print("total vowel come in it is ", count)
```

#### 4. WAP to enter a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left.  
<h4>Sample Input <br> 
ABCDCDC  <br>
CDC <br></h4>
 
<h4>Sample Output<br>
2 </h4>

```python
main_string=input("enter a string:")#banana
sub_string=input("enter a sub string:")#na
count=0
n=len(main_string)#6
m=len(sub_string)#2
for i in range(n-m+1):
    if(main_string[i:i+m]==sub_string):
        count+=1

print("",count)
```

#### 5. Given a string containing both upper and lower-case alphabets. Write a Python program to count the number of occurrences of each alphabet (case insensitive) and display the same.  
<h4>Sample Input <br> 
ABaBCbGc  <br></h4>

<h4>Sample Output <br> 
2A  <br>
3B  <br>
2C  <br>
1G </h4>

```python
stri=input("enter a string:")
a=stri.lower()
a=sorted(set(stri))
for idx in a:
    count=stri.count(idx)
    print(count,idx)
```

#### 6. Program to count the number of unique words in a given sentence using sets. 

```python
word=input("enter a word:")
m=word.upper()
se=sorted(set(word))
for w in se:
    co=word.count(w)
    print(co,w)
```

<h4> 7. Create 2 sets s1 and s2 of n fruits each by taking input from user and find: <br>  
 a. Fruits which are in both sets s1 and s2  <br>
 b. Fruits only in s1 but not in s2  <br>
 c. Count of all fruits from s1 and s2 </h4>

```python
n=int(input("enter a number: "))

s1=set()
s2=set()

print("Enter fruits for set1: ")
for i in range(n):
    fruit=input()
    s1.add(fruit)

print("Enter fruits for set2: ")
for i in range(n):
    fruit=input()
    s2.add(fruit)

print("Fruits in both s1 and s2:", s1 & s2)

print("Fruits only in s1:", s1-s2)

print("Total count of unique fruits:", len(s1|s2))
```

<h4> 8. Take two sets and apply various set operations on them:<br>  
 S1 = {Red, yellow, orange, blue}  <br>
 S2 = {violet, blue, purple} </h4>

```python
s1={"Red", "yellow", "orange", "blue"}
s2 = {"violet", "blue", "purple"}
print(s1.intersection(s2)) 
print(s1.union(s2))
print(s1-s2)
```
