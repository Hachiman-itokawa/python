<h4>1. Add few names, one name in each row, in “name.txt file”.<br>  
a. Count no of names<br>  
b. Count all names starting with vowel<br>
c. Find longest name  </h4>

```python

with open('name.txt', 'r') as file:
    names = [line.strip() for line in file if line.strip()]

num_names = len(names)
print(f"Number of names: {num_names}")

vowels = 'aeiouAEIOU'
vowel_count = sum(1 for name in names if name[0] in vowels)
print(f"Number of names starting with a vowel: {vowel_count}")

if names:
    longest_name = max(names, key=len)
    print(f"Longest name: {longest_name}")
else:
    print("No names found.")
```

<h4>2. Store integers in a file. <br> 
a. Find the max number  <br>
b. Find average of all numbers<br>  
c. Count number of numbers greater than 100 </h4>

```python

with open('numbers.txt', 'r') as file:
    numbers = [int(line.strip()) for line in file if line.strip()]


if numbers:
    max_num = max(numbers)
    print(f"Max number: {max_num}")
else:
    print("No numbers found.")


if numbers:
    avg = sum(numbers) / len(numbers)
    print(f"Average: {avg:.2f}")
else:
    print("No numbers found.")


count_gt_100 = sum(1 for num in numbers if num > 100)
print(f"Numbers greater than 100: {count_gt_100}")
```

<h4>3. Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) ):  <br>
Example:  <br>
Dehradun 5.78 308.20 <br> 
Delhi 190 1484  <br>
……………  <br>
Open file city.txt and read to:<br>  
a. Display details of all cities  <br>
b. Display city names with population more than 10Lakhs<br>  
c. Display sum of areas of all cities  </h4>

```python

cities = []
with open('city.txt', 'r') as file:
    for line in file:
        parts = line.strip().split()
        if len(parts) == 3:
            city, pop, area = parts
            cities.append((city, float(pop), float(area)))


print("Details of all cities:")
for city, pop, area in cities:
    print(f"{city}: Population {pop} lakhs, Area {area} sq KM")


print("\nCities with population more than 10 lakhs:")
for city, pop, area in cities:
    if pop > 10:
        print(city)

total_area = sum(area for city, pop, area in cities)
print(f"\nSum of areas: {total_area} sq KM")
```

<h4>4. Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError. <br>  
Sample input<br>  
1 0  <br>
2 $  <br>
3 1   <br>
Sample Output:<br>  
Error Code: integer division or modulo by zero   <br>
Error Code: invalid literal for int() with base 10: '$' 3  </h4>

```python
inputs = [
    "1 0",
    "2 $",
    "3 1"
]

for line in inputs:
    try:
        a, b = line.split()
        a = int(a)
        b = int(b)
        result = a // b
        print(result)
    except ZeroDivisionError:
        print("Error Code: integer division or modulo by zero")
    except ValueError:
        print("Error Code: invalid literal for int() with base 10: '$'")
```

#### 5. Create multiple suitable exceptions for a file handling program. 

```python

import os


try:
    with open('nonexistent.txt', 'r') as f:
        content = f.read()
except FileNotFoundError:
    print("FileNotFoundError: The file does not exist.")

try:
    with open('city.txt', 'w') as f: 
        f.write("test")
    print("File written successfully.")
except PermissionError:
    print("PermissionError: No permission to write to the file.")


try:
    with open('.', 'r') as f:
        content = f.read()
except IOError:
    print("IOError: Is a directory")

try:
    with open('city.txt', 'r', encoding='utf-8') as f:
        content = f.read()
    
except UnicodeDecodeError:
    print("UnicodeDecodeError: Decoding failed.")

try:
    with open('city.txt', 'invalid_mode') as f:
        pass
except ValueError:
    print("ValueError: Invalid file mode.")

print("File handling exceptions demonstrated.")
```
