#### 9. Write a program to convert given seconds into hours, minutes and remaining seconds.

```python
s=int(input("enter a second: "))
hrs=s//3600
remaining_sec=s%3600
min=remaining_sec//60
sec=remaining_sec%60
print("hrs: ", hrs)
print("min: ", min)
print("sec: ", sec)
```
