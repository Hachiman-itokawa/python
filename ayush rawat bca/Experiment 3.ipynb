5. Check whether the quadratic equation has real roots or imaginary roots. 
Display the roots.  

```python
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))
c = int(input("Enter a third number: "))
d = b**2 - 4*a*c
if d > 0:
    print("The equation has two real roots", d)
elif d == 0:
    print("The equation has one real root", d)
elif d < 0:
    print("The equation has imaginary roots", d)
```
