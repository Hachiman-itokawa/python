3. Find the greatest among two numbers. If numbers are equal than print 
“numbers are equal”. 

```python
a=int(input("Enter a no:"))
b=int(input("Enter a sec no:"))
if a>b:
    print(f"{a} is greater than {b}")
elif b>a:
    print(f"{b} is greater than {a}")
elif a==b:
    print("both numbers are equal")
```
