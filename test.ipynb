```python
def missing_numbers(nums):
    a=len(nums)
    total=a*(a+1)//2
    return int(total-sum(nums))

nums=[3,0,1]
t=missing_numbers(nums)
print(t)
```

```python
def top_k(items,k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]

items=[("A", 90), ("B", 95),("c", 95),]
k=2
print(top_k(items,k))
```

```python
def count_by_category(pairs):
    d={}
    for cat,word in pairs:
        if cat not in d:
            d[cat]={}
        if word not in d[cat]:
            d[cat][word]=0
            d[cat][word]+=1
    return d

pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
p=count_by_category(pairs)
print=(p)
```
